### 3.1 为什么选取 $u = e^{\sigma \sqrt{\Delta t}},\,d = e^{-\sigma \sqrt{\Delta t}}$

---

我们的目标是使得二叉树模型逼近几何布朗运动（GBM），在 Black–Scholes 框架下，若资产 $S_t$ 满足随机微分方程  
$$
dS_t = \mu\,S_t\,dt + \sigma\,S_t\,dW_t,
$$
（具体证明见**3.2 几何布朗运动的显示解**）其解为：
$$
S_T = S_0 \exp\!\bigl((\mu - \tfrac{1}{2}\sigma^2)\,T + \sigma\,W_T\bigr).
$$
因此，$\ln(S_T)$ 在时间 $T$ 时刻服从正态分布，具体为  
$$
\ln(S_T)
\sim
N\Bigl(\ln S_0 + \bigl(\mu - \tfrac{1}{2}\sigma^2\bigr)T,\;\sigma^2\,T\Bigr).
$$
这意味着 $S_T$ 本身具有 **对数正态分布（lognormal）**, 特别地，在短时间$ \Delta t $内，有短期对数收益：
$$ \ln\!\bigl(S_{t+\Delta t}/S_t\bigr) \sim N\Bigl(\bigl(\mu - \tfrac12\,\sigma^2\bigr)\,\Delta t,\;\sigma^2\,\Delta t\Bigr), $$
即短期对数收益的方差等于$ \sigma^2 \Delta t. $  在离散时间的 **CRR 二叉树模型** 中，我们将区间 $[0,T]$ 分成 $N$ 步，每步长度 $\Delta t = T/N$。若要在 $N \to \infty$（$\Delta t \to 0$）时与 GBM 收敛一致，需要保证：  
- 单步对数收益的方差近似 $\sigma^2\,\Delta t$；  
- 多步累加后，对数收益能够趋于正态分布。

接下来我们设置 $u,d$ 使得方差匹配，在二叉树模型的每个时间步内，假设资产价格只可能：
- 以上升因子 $u$ 的倍数上升：$S_{t+\Delta t} = S_t \cdot u$
- 以下降因子 $d$ 的倍数下降：$S_{t+\Delta t} = S_t \cdot d$

若令
$$
u = e^a,\quad d = e^{-a},
$$
则对数收益只有 $\ln(u) = a$ 和 $\ln(d) = -a$ 两种，呈 $\pm a$ 的对称分布。

记单步上涨的概率为 $p$。则对数收益的期望与方差分别是：
$$
\begin{aligned}
E[\ln(S_{t+\Delta t}/S_t)]
&= p\,a + (1-p)\,(-a) 
= (2p - 1)\,a,\\
\mathrm{Var}[\ln(S_{t+\Delta t}/S_t)]
&= p\,a^2 + (1-p)\,a^2 - \bigl((2p-1)\,a\bigr)^2 
= a^2 \bigl[1 - (2p -1)^2\bigr].
\end{aligned}
$$

为了与 GBM 的方差 $\sigma^2\,\Delta t$ 相同，需要：
$$
a^2 \bigl[1 - (2p-1)^2\bigr]
= \sigma^2\,\Delta t.
$$

在 **CRR 模型** 中，常取对称概率 $p = \tfrac{1}{2}$，此时 $1 - (2p-1)^2 = 1$，可得
$$
a^2 = \sigma^2 \,\Delta t 
\quad\Longrightarrow\quad
a = \sigma \sqrt{\Delta t}.
$$
故
$$
\boxed{u = e^{\sigma \sqrt{\Delta t}}, \quad d = e^{-\sigma \sqrt{\Delta t}}.}
$$

因此，$u = e^{\sigma \sqrt{\Delta t}},\,d = e^{-\sigma \sqrt{\Delta t}}$是 CRR 二叉树中最典型的参数设定，用以确保离散化方案在极限情形下收敛到连续时间的几何布朗运动。

### 3.2 定理（几何布朗运动的显式解）

令 $\{W_t\}_{t\ge 0}$ 为标准布朗运动，考虑随机微分方程（SDE）
$$
dS_t \;=\; \mu\,S_t\,dt \;+\; \sigma\,S_t\,dW_t, 
\quad S_0>0.
$$
则该 SDE 存在唯一强解（strong solution），并且该解可显式写为
$$
S_T \;=\; S_0 \exp\!\Bigl[\bigl(\mu - \tfrac{1}{2}\,\sigma^2\bigr)\,T \;+\; \sigma\,W_T\Bigr].
$$

这是一个比较重要的结论，在后续章节中的蒙特卡洛模拟中也有对这个结论的应用。

---
#### 证明

1. **对数变换并应用 Itô 引理**

   为求解方程
   $$
   dS_t \;=\; \mu\,S_t\,dt \;+\; \sigma\,S_t\,dW_t,
   $$
   首先令
   $$
   X_t \;=\; \ln(S_t),
   $$
   因为 $S_t>0$ 保证了 $\ln(S_t)$ 的良定义, 在伊藤积分理论（Itô 公式）中，若对可微函数 $f(x)$ 应用 Itô 引理，且随机过程 $\{Y_t\}$ 满足
   $$
   dY_t \;=\; a_t\,dt \;+\; b_t\,dW_t,
   $$
   则
   $$
   df(Y_t) 
   \;=\; 
   \frac{\partial f}{\partial x}(Y_t)\,dY_t 
   \;+\; 
   \tfrac{1}{2}\,\frac{\partial^2 f}{\partial x^2}(Y_t)\,(b_t)^2\,dt.
   $$
   
   在此，我们对 $f(x)=\ln x$，过程 $Y_t=S_t$，故
   $$
   a_t \;=\; \mu\,S_t,
   \quad
   b_t \;=\; \sigma\,S_t.
   $$
   那么
   $$
   \frac{\partial}{\partial x}\ln x \;=\; \frac{1}{x},
   \quad
   \frac{\partial^2}{\partial x^2}\ln x \;=\; -\,\frac{1}{x^2}.
   $$
   因而由 Itô 公式可得
   $$
   dX_t
   \;=\;
   d(\ln S_t)
   \;=\;
   \frac{1}{S_t}\bigl(\mu\,S_t\,dt + \sigma\,S_t\,dW_t\bigr)
   \;-\;
   \tfrac12\,\sigma^2\,dt.
   $$
   其中最后一项的系数 $-\tfrac12\sigma^2$ 来自
   $$
   \tfrac12\,\Bigl(-\tfrac{1}{S_t^2}\Bigr)\,(\sigma\,S_t)^2 
   \;=\; 
   -\,\tfrac12\,\sigma^2.
   $$
   整理后得到
   $$
   dX_t
   \;=\;
   \bigl(\mu - \tfrac12\,\sigma^2\bigr)\,dt
   \;+\;
   \sigma\,dW_t.
   $$

2. **积分并求解**

   上式是关于 $X_t$ 的线性随机微分方程，可将其从 0 积分到 $T$：
   $$
   X_T - X_0
   \;=\;
   \int_0^T \Bigl(\mu - \tfrac12\sigma^2\Bigr)\,dt
   \;+\;
   \sigma \int_0^T dW_t.
   $$
   由标准布朗运动的性质，$\int_0^T dW_t = W_T - W_0$。通常取 $W_0=0$，于是
   $$
   X_T - X_0
   \;=\;
   \bigl(\mu - \tfrac12\sigma^2\bigr)\,T
   \;+\;
   \sigma\,W_T.
   $$
   又因为 $X_0 = \ln S_0$，故
   $$
   X_T
   \;=\;
   \ln S_0
   \;+\;
   \bigl(\mu - \tfrac12\sigma^2\bigr)\,T
   \;+\;
   \sigma\,W_T.
   $$

3. **指数化得到 $S_T$**

   将上式两边指数化，就得到了过程 $\{S_T\}$ 的显式解：
   $$
   S_T
   \;=\;
   e^{X_T}
   \;=\;
   \exp\!\Bigl[
       \ln S_0
       \;+\;
       \bigl(\mu - \tfrac12\sigma^2\bigr)\,T
       \;+\;
       \sigma\,W_T
   \Bigr]
   \;=\;
   S_0 \,\exp\!\Bigl[
       \bigl(\mu - \tfrac12\sigma^2\bigr)\,T
       \;+\;
       \sigma\,W_T
   \Bigr].
   $$

   至此，我们完成了对几何布朗运动显式解的推导。

4. **唯一性与正值性**

   - 在一般随机微分方程理论（特别是 Lipschitz 条件可满足时）可证得解的**存在唯一性**。具体地，本问题中漂移项 $\mu S_t$ 与扩散项 $\sigma S_t$ 均为线性的，满足全局 Lipschitz 条件，因此存在唯一强解。  
   - 由上述显式解形式可以看出 $\exp(\cdot)$ 始终大于 0，因而 $S_t>0$ 对所有 $t$ 成立。

综上，
$$
S_T 
\;=\;
S_0 \,\exp\!\Bigl[
    \bigl(\mu - \tfrac12\sigma^2\bigr)\,T
    \;+\;
    \sigma\,W_T
\Bigr]
$$
即为所给几何布朗运动（Geometric Brownian Motion）的唯一强解，证毕。

---

**注记**：上述 $-\tfrac12\sigma^2$ 项对应伊藤积分中的“二次变差”修正，是区别于经典微分方程 $\tfrac{d}{dt}\ln S_t$ 的关键因素，也是 Black–Scholes 公式等衍生品定价理论中不可或缺的部分。
