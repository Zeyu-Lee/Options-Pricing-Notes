## 引言

### 1. 金融衍生品

衍生品是一类金融合约或金融工具，其价值源自某一标的资产。该标的资产可以是可交易资产（如股票）、指数组合、期货价格、商业地产，甚至是某种可度量的状态变量（如特定地点的天气状况）。衍生品的收益结构多种多样，既可以是均匀分布于整个时间周期的现金流，也可以在特定日期发生，或是两者的结合。因此，衍生品也被称为**或有权证（Contingent Claims）** 。

衍生品主要包括**远期（Forwards）、期货（Futures）、期权（Options）和互换（Swaps）**，在本模块中笔者主要对各种期权的不同定价方法进行探讨 。

### 2. 期权

期权是一种由买卖双方签订的合约，它赋予买方在未来某个时间以约定价格买入或卖出标的资产的权利，但不具有必须执行的义务。期权主要分为两类：**看涨期权（Call Option）和看跌期权（Put Option）** 。

* **看涨期权**（Call Option）：赋予持有人以约定价格购买标的资产的权利；
* **看跌期权**（Put Option）：赋予持有人以约定价格出售标的资产的权利。

合约中规定的交易价格称为**执行价格（Strike Price）** ，合约到期的时间称为**到期日（Expiry Date）** 。

市场上大多数期权属于**欧式期权（European Option）或美式期权（American Option）** ，除此之外，还有**障碍期权（Barrier Option）、百慕大期权（Bermudan Option）、亚式期权（Asian Option）** 及 **回望期权(Lookback Option)** 等特殊类型。

### 3. 欧式期权与美式期权

欧式期权是最基本的期权类型。其特点是**只能在到期日行权**，其收益计算公式如下：

* 欧式看涨期权（Call Option）：$\Phi = \max(S(T) - K, 0)$
* 欧式看跌期权（Put Option）：$\Phi = \max(K - S(T), 0)$

其中，$S(T)$ 表示标的资产在到期日 $T$ 的价格，$K$ 为执行价格。

相比之下，美式期权的行权方式更加灵活，**可在任意时间点行权**，即持有人可以选择在到期日之前的任何时间执行期权。因此，其收益计算如下：

* 美式看涨期权（Call Option）：$\Phi = \max_{t \leq \tau \leq T} (S(\tau) - K, 0)$
* 美式看跌期权（Put Option）：$\Phi = \max_{t \leq \tau \leq T} (K - S(\tau), 0)$

其中，$\tau$ 代表行权时间。

显然，由于美式期权可以在任意时间点行权，因此其定价和分析比欧式期权更加复杂，也更具研究价值。

### 4. 期权的用途

期权主要有两个核心用途：**投机（Speculation）和对冲（Hedging）** 。

 **（1）投机**：
相较于直接投资标的资产，期权投资具有杠杆效应，因此投机者可以利用期权实现更高的收益。例如，如果投资者认为爱立信（Ericsson）股票即将上涨，他可以购买相应的看涨期权，以较低的初始投入获取高额回报。因此，期权在金融市场上越来越受到投资者青睐。

 **（2）对冲**：
期权也是一种有效的风险管理工具，尤其对于期权卖方而言，他们通常希望通过对冲策略规避潜在的巨大风险。例如，期权卖方可以通过出售期权来获得一定的权利金收入，并利用其他金融工具对冲潜在损失，从而锁定收益。这种无风险套利思想在期权定价理论与实践中占据核心地位。

### 5. 期权定价

尽管期权的概念自古已有，但直到 **Black-Scholes(1973)** 期权定价公式提出后，才形成了一套理论完备的定价框架。如今，期权定价已成为金融市场研究的重要组成部分。

由于**Black-Scholes模型**的适用范围较为有限，后来又发展出了多种更复杂的定价方法，包括以下三种主流方法：

* **二叉树（Binomial Methods）** ：
  该方法通过构造期权价格的**二叉树**，在每个时间节点计算期权的理论价值。利用**无风险套利**思想，可在每个节点计算期权价格，从而得出最终定价。相较于Black-Scholes模型，二叉树法的优势在于更具灵活性，例如可正确建模**离散现金流**（如股息支付）及**美式期权**。
* **有限差分法（Finite Difference Methods）** ：
  若期权定价可以表示为**偏微分方程（PDE）** ，则可通过**有限差分法**求解。该方法利用**泰勒展开(Taylor Expansion)** 对偏微分方程的导数项进行离散近似求解。需要注意的是，二叉树方法实际上是 **显式有限差分法（Explicit Finite Difference Methods)** 的特例。
* **蒙特卡洛方法（Monte Carlo Methods）** ：
  对于结构复杂、无法通过传统解析方法求解的期权（如路径依赖型期权），蒙特卡洛模拟是一种有效的近似求解工具。该方法通过随机模拟大量可能的价格路径，并计算期权的平均收益，以此得到期权的理论价值。
